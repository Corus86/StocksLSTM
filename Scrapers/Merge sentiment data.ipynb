```python
import pandas as pd
```

```python
df = pd.read_csv('Summanry_db.csv')
```

```python
df_apple = pd.read_csv('Apple Sentiment.csv')
```

```python
df_apple
```

```python
df_subapp = df[df.get('Company') == 'Apple']
```

```python
merged_Apple = pd.merge(df_subapp, df_apple, left_on = 'formatted_date', right_on = 'Tweet Datetime')
```

```python
merged_Apple
```

```python
merged_Apple
```

```python
merged_Apple.to_csv('Apple merged.csv')
```
